### Correlation

Relationship between two data columns or two attributes/features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../Data/bodyfat.csv")
```

```python
df.head()
```

```python
df.corr()
```

corr =-1 and 1<br>
if corr value is near to 1 -> positive correlation <br>
if corr value is equal to 0 -> there is no relation ship / neural<br>
if corr value is close to -1 -> negative correlation<br>

```python
fig=plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

```python
df=pd.DataFrame({"Data":["red","green","blue"]})
df
```

```python
#scikit-learn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
#fit
#transform
#inverse_transform
#fit_transform
```

```python
#fit
le.fit(df["Data"])
```

```python
le.classes_
```

```python
le.transform(df["Data"])
```

```python
le.transform(["green"])
```

```python
le.inverse_transform([0,1])
```

```python
le.fit_transform(df["Data"])
```

### One hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder()
#fit
#transform
#inverse_transform
#fit_transform
```

```python
oh.fit(df[["Data"]])
```

```python
oh.transform(df[["Data"]]).toarray()
```

```python
oh.inverse_transform([[0., 0., 1.]])
```

```python
oh.fit_transform(df[["Data"]]).toarray()
```
